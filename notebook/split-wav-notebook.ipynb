### Split .wav and save

```python
import librosa
import numpy as np
from tqdm import tqdm
import glob
from multiprocessing import Pool
import soundfile as sf
```

```python
DATA_DIR = '/home/elicer/GuitarSet/audio_mono-mic/'
SAVE_DIR = '/home/elicer/GuitarSet/audio_mono-mic-split/'
file_list = glob.glob(DATA_DIR +'*solo*.wav')
```

```python
f = file_list[0]
duration = 5 *22050
```

```python
for f in file_list:
    sig, sr = librosa.load(f)
    fn = f.split('/')[-1]
    fn_ = fn.split('.wav')[0]
    for i in range(len(sig)//duration):
        chunk = sig[i*duration: (i+1) * duration]
        sf.write( SAVE_DIR + f"{fn_}_{i}.wav", data=chunk, samplerate=sr)
        print('saved')
    
    chunk = sig[(i+1) * duration:]
    if len(chunk) >= (5*22050):
        sf.write( SAVE_DIR + f"{fn_}_{i+1}.wav", data=chunk, samplerate=sr)
        print('saved')
```
